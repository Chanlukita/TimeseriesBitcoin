```python
import pandas as pd
import numpy as np
from scipy import stats

df = pd.read_csv('BTC_USD_2018-01-01_2020-12-31-CoinDesk.csv')

# Fungsi untuk menghitung Z-score dan menghapus outliers
def remove_outliers_zscore(df, columns, threshold=3):
    z_scores = np.abs(stats.zscore(df[columns]))
    filter = (z_scores < threshold).all(axis=1)
    return df[filter]

# Menerapkan Z-score handling pada kolom yang relevan
columns_to_check = ['Closing Price (USD)', '24h Open (USD)', '24h High (USD)', '24h Low (USD)']
df_cleaned = remove_outliers_zscore(df, columns_to_check)

# Simpan DataFrame yang telah dibersihkan ke CSV
df_cleaned.to_csv('ZScore.csv', index=False)

print("Data setelah menghapus outliers disimpan ke ZScore.csv")
print(df_cleaned)
```
